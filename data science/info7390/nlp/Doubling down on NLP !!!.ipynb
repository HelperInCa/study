# Let's go grab a webpage and ask our NLP robot to tell us what's it's all about !

```python
my_page = urllib.request.urlopen('https://en.wikipedia.org/wiki/Warren_Buffett')
my_page_html = my_page.read()
```

```python
'''What did we just grab ?'''
print(my_page_html)
```

# Well, how about we talk beautiful soup here?

```python
'''From soup to beautiful soup...heheh'''

soup = BeautifulSoup(my_page_html,'html5lib')
beau_soup = soup.get_text(strip = True)
print(beau_soup)
```

# Tokens, tokenize...what?

```python
tokens = list()
```

```python
for token in beau_soup.split():
    tokens.append(token)
print(tokens)
```

# Well, what is it trying to tell us ?

```python
'''The stop words we were talking about ?Let's remove them !'''

remov_words = stopwords.words('english')
better_tokens = tokens[:]

for token in tokens:
    if token in stopwords.words('english'):
        better_tokens.remove(token)
```

```python
'''Off to calcualte word frequency assuming the most frequent word will tell us what this story is all about'''

word_count = nltk.FreqDist(better_tokens)

'''word_count is a python dictionary, ok'''

for key,val in word_count.items():
    print(str(key) + ':' + str(val))
```

```python
'''How about visualizing it'''

plt.figure(figsize = (10,10))
word_count.plot(10,cumulative=False)
```
