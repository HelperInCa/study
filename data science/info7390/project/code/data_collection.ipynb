```python
import numpy as np

new_principal = pd.DataFrame(columns = ['tconst', 'actor', 'actress', 'archive_footage', 'archive_sound', \
                'cinematographer', 'composer', 'director', 'editor', 'producer', 'production_designer', \
                'self', 'writer'])
# print(new_principal.info())

df = principal

filter_movie_title = pd.merge(movie_title, rating, on = ['tconst'], how = 'inner')

title_list =  filter_movie_title.loc[:, 'tconst'].tolist()

count = 0
pre_percent = 0

print('title_list', len(title_list))
print('filter_movie_title', len(filter_movie_title))

for tconst in title_list:
    tmp_df = df[df.tconst == tconst]
    
    percent = round((count / len(title_list)) * 100, 2)
    if percent != pre_percent:
        print(percent, '%')
        pre_percent = percent
        
    dict = {'tconst': tconst, 'actor':'', 'actress':'', 'archive_footage':'', 'archive_sound':'', \
            'cinematographer':'', 'composer':'', 'director':'', 'editor':'', 'producer':'', \
            'production_designer':'', 'self':'', 'writer':''}
            
    count = count + 1
        
    for i, row in tmp_df.iterrows():
        cur_category = row['category']
        cur_name = row['nconst']
                
        if dict[cur_category] != '' :
            dict[cur_category] += ','
        dict[cur_category] += cur_name
               
    tmp = pd.DataFrame([dict])
    new_principal = new_principal.append(tmp)
            
print(new_principal.info())
print(new_principal.head(5))
```

```python
df = pd.merge(filter_movie_title, new_principal, on = ['tconst'], how = 'inner')
print(df.info())
print(df.head)

df.to_csv("/Users/elias/Desktop/result2010-2020.csv", index = False, header=True)
```
